## OneHotEncoder -->similar to pd.get_dummies

- Import the OneHotEncoder Class
- Create an object for the class
- ColumnTransformer--> Insider this columntransform, we will pass this OneHotEncoder
- It has two mandatory arguments --> transforms,remainder

```python
##Importing Libraries
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Insider the columntransformer, we have to pass the OneHotEncoder
column_transformer=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
### syntax:objectname.fit_transform(data)
```

```python
data =column_transformer.fit_transform(data)
data
```

```python
#creating a dataframe out of array
X=pd.DataFrame(data)
X
```

```python
### importing the classes
### creating the objects
### To apply the object to the dataframe(columns),we are using fit_transform --> With help of fit_transform method
### we are encoding the categorical variables
### LabelEncoder  is used only when we have to classes
```

## Feature scaling:(Not mandatory for all the algorithms)
- scale()
- MinMaxscaler()
- Standardscaler()

```python
sample = pd.DataFrame({'W.E':[2,4,5,6,7,8],'Age':[28,25,32,45,33,32],'Salary':[300000,400000,500000,600000,700000,800000],
                      'Distance':[10,12,23,32,15,17],'Expected Salary':[400000,500000,600000,700000,800000,900000]})
sample
```

- Most the Machine Learning Algorithms,we use distance based formulas
* Salary column is hugely dominating the other columns in terms of magnitude
* Also,all the columns in our dataset,belong to different units,Exp-->Years, Salary-->Currency
* Units and Scale(magnitude)
* Feature Scaling--> when applied, this units and scaling will be removed
* To make the data unitless and scaless, we have to apply Feature Scaling
