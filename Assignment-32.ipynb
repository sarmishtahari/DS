```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Youtube05-Shakira.csv")
data.head()
```

```python
data.CLASS.value_counts()
```

```python
data.drop(["COMMENT_ID","AUTHOR","DATE"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
plt.title("count of ham's and spam's in dataframe")
sns.set(style="darkgrid")
sns.countplot(data.CLASS,palette=["red","blue"])
plt.legend
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.CONTENT,data.CLASS,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv
```

```python
word_cv=cv.fit(X_train)
word_cv.get_feature_names()
```

```python
doc_arr=word_cv.transform(X_train).toarray()
doc_arr
```

```python
feature_mat=pd.DataFrame(doc_arr,columns=word_cv.get_feature_names())
feature_mat
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid=TfidfVectorizer()
```

```python
X_train_trans=tfid.fit_transform(X_train)
X_train_trans
```

```python
X_test_trans=tfid.transform(X_test)
X_test_trans
```

```python
tfid.get_feature_names()
```

```python
tfid.vocabulary_
```

```python
tfid.idf_
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb=BernoulliNB()
```

```python
bnb.fit(X_train_trans,y_train)
```

```python
y_pred=bnb.predict(X_test_trans)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
