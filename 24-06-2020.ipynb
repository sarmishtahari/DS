####  Out of 13 columns--> today we will work with
#### Single input and medv as output

#### we are using RM columns as input (SLR--> we have only one input)
#### output-->MEDV column

```python
# Preparing the Input and Output columns
X=boston_df[['RM']] #selecting dataframe columns
y=boston_df['MEDV']
```

```python
boston_df['MEDV'] #selecting columns as series
```

```python
#plt.style.use
with plt.style.context("grayscale"):
    plt.plot(X,y,"bo")
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("Scatter Plot")
```

### Applying Linear Regression on data:

```python
## Usual sklearn:import,instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,y)
```

```python
##Print intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
# y=mx+c
#m value--> coefficient(9.10210898)
#c value--> intercept(-34.67062078)
(9.10210898 * 6.575) + (-34.67062078)
```

## Interpreting the model:
    A unit increase in X(RM) is associated with  a 9.10210898 unit increase in MEDV

#### Pipeline-->First we create a model on the data
#### Second, we need to compare the model(predicted values with actual values)

```python
## with the help of predict,we are going to find our predicted values
y_pred=lm.predict(X)
y_pred
```

```python
#compare  the actual values with the predict values 
y==y_pred
```

```python
## r2_value--> to check the model performance
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
##observe the regression line 
with sns.axes_style('dark'):
    sns.regplot(X,y)
```

```python
## Scatter plot(Regression)
#First we are making a scatterplot of the original data()X and Y).
#we need to plot a line for thepredicted values(X,y_pred)
plt.scatter(X,y,color="red")
plt.plot(X,y_pred,color="black")
plt.xlabel("RM I/P")
plt.ylabel("MEDV O/P")
```
