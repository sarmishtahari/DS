#### Create a dataframe by imputing the missing values with mean Create a dataframe by imputing the missing values with median

```python
data.isna().sum()
```

```python
data1=data.copy()
data1
```

```python
df=pd.DataFrame(data1)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
df.isna().sum()
```

### If any missing values in categorical columns, impute it with mode

```python
df=pd.DataFrame(data1)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
df
```

```python
df.isna().sum()
```

### For the categorical columns apply different kinds of encoding techniques

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Sex=le.fit_transform(df.Sex)
df.Name=le.fit_transform(df.Name)
df.Ticket=le.fit_transform(df.Ticket)
df.Cabin=le.fit_transform(df.Cabin)
df.Embarked=le.fit_transform(df.Embarked)
df
```
