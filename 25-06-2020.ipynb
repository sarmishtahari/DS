```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the dataset from sklearn library
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston_df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV']=boston.target
boston_df.head()
```

```python
##To find the correlation we use corr()
boston_df.corr()
```

```python
## Visualizing the correlation
sns.heatmap(boston_df.corr())
```

```python
## Visualizing the correlation
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')
```

```python
## Split the data into training and testing;Build the model on training data and test the model on testing data
```

```python
#Split the data into input and output
X=boston_df.drop(['MEDV'],axis=1)#Input columns
y=boston_df.MEDV#boston_df['MEDV']
```

```python
#Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Build the model on Training data-->X_train and y_train
#sklearn library:import ,instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#Intercept and the coefficient values
print(lm.intercept_)
lm.coef_
```

```python
#create a dataframe for coefficients
coefficients=pd.DataFrame([X_train.columns,lm.coef_])
coefficients
```

```python
coefficients=coefficients.T
coefficients
```

```python
# y = b0+b1X1+b2X2+.......b13X13
# y = 36.93325545711978 + (-0.117735)()
```

```python
##Checking the model prediction on training data 
y_train_pred = lm.predict(X_train)
y_train_pred
```

```python
# We need to compare the actual values (y_train) and the predicted #values (y_train_pred)
from sklearn.metrics import r2_score
print('R^2',r2_score(y_train,y_train_pred))
```

```python
# We need to compare the actual values (y_train) and the predicted #values (y_train_pred)
from sklearn.metrics import r2_score
print('R^2',r2_score(y_train,y_train_pred))
print('Adjusted R^2:', 1- (1-r2_score(y_train, y_train_pred))*(len(X_train)-1)/
                                  (len(X_train)-X_train.shape[1]-1))
```

```python
numerator = (1-r2_score(y_train_pred,y_train))*(len(X_train)-1)

denominator = (len(X_train)-X_train.shape[1]-1)
print("Adjusted R^2",1- (numerator/denominator))
```

```python
## Visualize the differences between actual values and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model Predicted Price')
```

```python
## y_train-y_train_pred==Errors
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel("Predicted Values")
plt.ylabel("residuals")
```

```python
## Test the model on testing data
y_test_pred=lm.predict(X_test) # test data--> unseen data
y_test_pred
```

```python
## We need to compare the actual values(y_test) and the predicted
#values(y_test_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('R^2: ',r2_score(y_test,y_test_pred))
print("Adjusted R^2 value:",1-(1-r2_score(y_test,y_test_pred))*(len(X_test)-1)/(len(X_test)-X_test.shape[1]-1))
print('MAE:',mean_absolute_error(y_test,y_test_pred))
print('MSE:',mean_squared_error(y_test,y_test_pred))
print('RMSE',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
