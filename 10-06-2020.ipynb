### Getting numerical columns

```python
tips.dtypes!='category'
```

```python
num_col= tips.columns[tips.dtypes!='category']
```

```python
val =1
for i in range(len(num_col)):
    for j in range(len(num_col)):
        plt.subplot(3,3,val)
        plt.scatter(tips[num_col[i]],tips[num_col[j]])
        plt.title(num_col[i]+'_'+num_col[j])
        val+=1
```
